```python
from difflib import SequenceMatcher

# Original list of book titles
book_titles = [
    "A Lineage Of Grace", "Lineage of Grace", "Atonement Child", "The Atonement Child",
    "Good Morning Holy Spirit", "Good morning Holy Spirit",
    "Am I Being Fooled", "Am I being fooled?", "Am I Being Fooled?",
    "Sell Like Crazy", "Sell like crazy",
    "12 week MBA", "12-Week MBA",
    "The Richest Man In Babylon", "The Richest Man in Babylon",
    "Atomic Habit", "Atomic Habits",
    "The Diary of A CEO", "The Diary of a CEO",
    "The Practice of God''s Presence", "The Practice of The Presence of God",
    "Psychology of Money", "The Psychology of Money",
    "No Excuses", "No Excuses (Power of Self Discipline)", "The Power of Self Discipline (No Excuses)"
]

def find_similar_books(titles, threshold=0.95):
    groups = []
    used = set()

    for i, title1 in enumerate(titles):
        if title1 in used:
            continue
        group = [title1]
        for j, title2 in enumerate(titles):
            if i != j and title2 not in used:
                similarity = SequenceMatcher(None, title1.lower(), title2.lower()).ratio()
                if similarity >= threshold:
                    group.append(title2)
                    used.add(title2)
        used.add(title1)
        groups.append(group)
    return groups

def merge_duplicates(groups):
    corrected = []
    for group in groups:
        canonical = sorted(group, key=lambda x: (-len(x), x))[0]  # pick longest version as canonical
        corrected.append(canonical)
    return sorted(set(corrected))

similar_groups = find_similar_books(book_titles, threshold=0.95)
cleaned_titles = merge_duplicates(similar_groups)

print("✅ Cleaned Titles:")
for title in cleaned_titles:
    print("-", title)
```

```python
# Copy original data
discount_data = data.copy()

# Apply 10% discount on selling price
discount_data['discounted_price'] = discount_data['selling_price'] * 0.90  # 10% off

# Recalculate revenue assuming the same quantity sold
discount_data['discounted_revenue'] = discount_data['discounted_price'] * discount_data['quantity']

# Recalculate profit
discount_data['discounted_profit'] = discount_data['discounted_revenue'] - discount_data['cost_price']

# Recalculate percent profit
discount_data['discounted_percent_profit'] = (discount_data['discounted_profit'] / discount_data['discounted_revenue']) * 100

# Summarize the totals
original_revenue = data['selling_price'].sum()
original_profit = data['profit'].sum()

discounted_revenue = discount_data['discounted_revenue'].sum()
discounted_profit = discount_data['discounted_profit'].sum()

# Print comparison
print(f"Original Total Revenue: ₦{original_revenue:,.2f}")
print(f"Original Total Profit: ₦{original_profit:,.2f}")
print()
print(f"Discounted Total Revenue: ₦{discounted_revenue:,.2f}")
print(f"Discounted Total Profit: ₦{discounted_profit:,.2f}")

# Optional: Create a DataFrame to visualize before vs after
summary = pd.DataFrame({
    'Scenario': ['Original', '10% Discount'],
    'Total Revenue': [original_revenue, discounted_revenue],
    'Total Profit': [original_profit, discounted_profit]
})

# Plot
summary.set_index('Scenario')[['Total Revenue', 'Total Profit']].plot(
    kind='bar',
    figsize=(8, 5),
    title='Effect of 10% Discount on Revenue and Profit',
    color=['skyblue', 'orange']
)
plt.ylabel('₦ Amount')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Create a list of discount percentages
discount_rates = [0, 5, 10, 15, 20]

# Create an empty list to store the results
discount_results = []

# Loop through each discount rate
for rate in discount_rates:
    discount_multiplier = (100 - rate) / 100  # e.g., for 10%, multiplier = 0.9
    
    # Calculate discounted price
    discounted_price = data['selling_price'] * discount_multiplier
    
    # Calculate new revenue and profit
    discounted_revenue = discounted_price * data['quantity']
    discounted_profit = discounted_revenue - data['cost_price']
    
    # Summarize totals
    total_revenue = discounted_revenue.sum()
    total_profit = discounted_profit.sum()
    
    # Append results
    discount_results.append({
        'Discount %': rate,
        'Total Revenue': total_revenue,
        'Total Profit': total_profit
    })

# Create a summary DataFrame
discount_summary = pd.DataFrame(discount_results)

# Display
print(discount_summary)

# Plotting
discount_summary.set_index('Discount %')[['Total Revenue', 'Total Profit']].plot(
    kind='line',
    marker='o',
    figsize=(8, 5),
    title='Revenue and Profit Across Different Discount Rates',
    color=['royalblue', 'darkorange']
)
plt.ylabel('₦ Amount')
plt.grid(True)
plt.xticks(discount_rates)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Example monthly profit data
monthly_profit = {
    'month': ['January 2025', 'February 2025', 'March 2025', 'April 2025', 'May 2025', 'June 2025'],
    'profit': [4500, 4700, 4200, 3900, 3100, 2500]  # Replace with your real monthly profits
}
monthly_profit = pd.DataFrame(monthly_profit)

# Convert month names to datetime
monthly_profit['month'] = pd.to_datetime(monthly_profit['month'], format='%B %Y')

# Prepare the data
X = np.arange(len(monthly_profit)).reshape(-1, 1)  # 0,1,2,3,4,5
y = monthly_profit['profit'].values

# Use Polynomial Regression
degree = 2
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

# Predict future profits
future_months = 6
X_future = np.arange(len(monthly_profit), len(monthly_profit) + future_months).reshape(-1, 1)
X_future_poly = poly.transform(X_future)
future_profits = model.predict(X_future_poly)

# Evaluate model
y_pred = model.predict(X_poly)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Model performance:")
print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")

# Create future months dates
last_month = monthly_profit['month'].max()
future_dates = pd.date_range(start=last_month + pd.offsets.MonthBegin(1), periods=future_months, freq='MS')

# Save predictions into a dataframe
future_df = pd.DataFrame({
    'Future Month': future_dates.strftime('%B %Y'),
    'Predicted Profit': np.round(future_profits, 2)
})

print("\nFuture Sales Predictions:")
print(future_df)

# Combine old and future data for plotting
full_months = pd.concat([
    monthly_profit[['month', 'profit']].rename(columns={'month': 'Month', 'profit': 'Profit'}),
    pd.DataFrame({'Month': future_dates, 'Profit': future_profits})
]).reset_index(drop=True)

# 📈 Plot
plt.figure(figsize=(12,6))
plt.plot(full_months['Month'], full_months['Profit'], marker='o', linestyle='-', color='blue')
plt.axvline(x=last_month, color='red', linestyle='--', label='Prediction Start')
plt.title('Monthly Profit and Future Predictions')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.grid(True)
plt.legend(['Profit', 'Prediction Start'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Save to Excel
full_months.to_excel('future_profit_predictions.xlsx', index=False)
print("\n✅ Future profit predictions saved to 'future_profit_predictions.xlsx'")

```
