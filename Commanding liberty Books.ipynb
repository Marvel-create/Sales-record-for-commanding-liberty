```python
import pandas as pd

# Function to clean money values like "₦1,700" into integers
def clean_money(value):
    if isinstance(value, str):
        return int(value.replace("₦", "").replace(",", "").strip())
    return value
```

```python
november = pd.read_csv(r"C:\Users\HP\Desktop\CLB\COMMANDING LIBERTY BOOKS (Finance Records NOV. 2024).csv")
december = pd.read_csv(r"C:\Users\HP\Desktop\CLB\COMMANDING LIBERTY BOOKS (Finance Records DEC 2024).csv")
january = pd.read_csv(r"C:\Users\HP\Desktop\CLB\COMMANDING LIBERTY BOOKS (Finance Records JAN 2025).csv")
febuary = pd.read_csv(r"C:\Users\HP\Desktop\CLB\COMMANDING LIBERTY BOOKS (Finance Records FEB 2025).csv")
march = pd.read_csv(r"C:\Users\HP\Desktop\CLB\COMMANDING LIBERTY BOOKS (Finance Records MARCH 2025).csv")
april = pd.read_csv(r"C:\Users\HP\Desktop\CLB\COMMANDING LIBERTY BOOKS (Finance Records APRIL 2025).csv")
```

```python
december.columns
```

```python
november.columns = november.columns.str.strip().str.lower().str.replace(" ", "_")
november["month"] = "November 2024"
november["cost_price"] = november["cost_price_(in_naira)"].apply(clean_money)
november["selling_price"] = november["selling_price_(in_naira)"].apply(clean_money)
november["profit"] = november["profit"].apply(clean_money)
november["percent_profit"] = november["%_profit"]

december.columns = december.columns.str.strip().str.lower().str.replace(" ", "_")
december["month"] = "December 2024"
december["cost_price"] = december["cost_price_(in_naira)"].apply(clean_money)
december["selling_price"] = december["selling_price_(in_naira)"].apply(clean_money)
december["profit"] = december["profit"].apply(clean_money)
december["percent_profit"] = december["%profit"]

january.columns = january.columns.str.strip().str.lower().str.replace(" ", "_")
january["month"] = "January 2025"
january["cost_price"] = january["cost_price_(in_naira)"].apply(clean_money)
january["selling_price"] = january["selling_price_(in_naira)"].apply(clean_money)
january["profit"] = january["profit"].apply(clean_money)
january["percent_profit"] = january["%_profit"]

febuary.columns = febuary.columns.str.strip().str.lower().str.replace(" ", "_")
febuary["month"] = "Febuary 2025"
febuary["cost_price"] = febuary["cost_price_(in_naira)"].apply(clean_money)
febuary["selling_price"] = febuary["selling_price_(in_naira)"].apply(clean_money)
febuary["profit"] = febuary["profit"].apply(clean_money)
febuary["percent_profit"] = febuary["%profit"]

march.columns = march.columns.str.strip().str.lower().str.replace(" ", "_")
march["month"] = "March 2025"
march["cost_price"] = march["cost_price_(in_naira)"].apply(clean_money)
march["selling_price"] = march["selling_price_(in_naira)"].apply(clean_money)
march["profit"] = march["profit"].apply(clean_money)
march["percent_profit"] = march["%profit"]

april.columns = april.columns.str.strip().str.lower().str.replace(" ", "_")
april["month"] = "April 2025"
april["cost_price"] = april["cost_price_(in_naira)"].apply(clean_money)
april["selling_price"] = april["selling_price_(in_naira)"].apply(clean_money)
april["profit"] = april["profit"].apply(clean_money)
april["percent_profit"] = april["%profit"]
```

```python
common_columns = [
    "books_sold", "authors", "name", "phone_number", 
    "cost_price", "selling_price", "profit", "percent_profit", "month"
]


merged_data = pd.concat([
    november[common_columns],
    december[common_columns],
    january[common_columns],
    febuary[common_columns],
    march[common_columns],
    april[common_columns],
], ignore_index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by month and calculate aggregates
monthly_summary = data.groupby('month')[['selling_price', 'profit', 'percent_profit', 'cost_price']].agg({
    'selling_price': 'sum',
    'profit': 'sum',
    'percent_profit': 'mean',
    'cost_price': 'sum'
}).reset_index()

# Fix typo in month name
monthly_summary['month'] = monthly_summary['month'].replace('Febuary 2025', 'February 2025')

# Convert month column to datetime format
monthly_summary['month'] = pd.to_datetime(monthly_summary['month'], format='%B %Y')

# Sort by month chronologically
monthly_summary = monthly_summary.sort_values('month')

# Line chart for revenue and profit
plt.figure(figsize=(12, 6))
plt.plot(monthly_summary['month'], monthly_summary['selling_price'], label='Revenue', marker='o')
plt.plot(monthly_summary['month'], monthly_summary['profit'], label='Profit', marker='o')
plt.plot(monthly_summary['month'], monthly_summary['cost_price'], label='Cost', marker='o')
plt.title("Monthly Revenue and Profit Trend")
plt.xlabel("Month")
plt.ylabel("Amount (₦)")
plt.legend()
plt.grid(True)
plt.show()

# Line chart for average profit margin
plt.figure(figsize=(12, 5))
plt.plot(monthly_summary['month'], monthly_summary['percent_profit'], label='Average Profit Margin (%)', color='green', marker='s')
plt.title("Monthly Average Profit Margin")
plt.xlabel("Month")
plt.ylabel("Average % Profit")
plt.grid(True)
plt.show()
```
